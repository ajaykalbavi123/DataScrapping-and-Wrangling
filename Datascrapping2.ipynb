```python
import bs4
import pandas as pd
import numpy as np
import requests
```

```python
monster_url = "https://www.monsterindia.com/data-science-jobs.html"
page = requests.get(monster_url).text
bs_obj = bs4.BeautifulSoup(page)
bs_obj
```

```python
job_parents = bs_obj.findAll('div',{'class':'jobwrap',
                                   'type':'tuple'})
len(job_parents)
```

```python
jobs = pd.DataFrame()
parents = bs_obj.findAll('div',{'class':'jobwrap','type':'tuple'})

for parent in job_parents:
    job_title = parent.find('span',{'class':'title_in'})
    company = parent.find('div',{'class':'jtxt orange'})
    location_tag = parent.find('div',{'class':'jtxt jico ico1'})
    if location_tag is None:
        location = None
    else:
        location = location_tag.text
    keyskills1 = parent.find('div',{'class':'joblnk serachjoblnk'})
    keyskills2 = keyskills1.findAll('div',{'class':'jtxt'})[3]
    keyskills = keyskills2.findAll('span')[1].text
    summary1 = parent.find('div',{'class':'joblnk serachjoblnk'})
    try:
        summary2 = summary1.findAll('div',{'class':'jtxt'})[4]
        fsummary = summary2.findAll('span')[1].text
    except:
        fsummary = None
    print(fsummary)
    print('-------')
```

```python
jobs = pd.DataFrame()
for i in range (1,5):
    monster_url = "https://www.monsterindia.com/data-science-jobs - %d.html" % i
    parents = bs_obj.findAll('div',{'class':'jobwrap','type':'tuple'})
    for parent in job_parents:
        job_title = parent.find('span',{'class':'title_in'}).text
        company = parent.find('div',{'class':'jtxt orange'}).text
        location_tag = parent.find('div',{'class':'jtxt jico ico1'})
        if location_tag is None:
            location = None
        else:
              location = location_tag.text
        keyskills1 = parent.find('div',{'class':'joblnk serachjoblnk'})
        keyskills2 = keyskills1.findAll('div',{'class':'jtxt'})[3]
        keyskills = keyskills2.findAll('span')[1].text
        summary1 = parent.find('div',{'class':'joblnk serachjoblnk'})
        try:
            summary2 = summary1.findAll('div',{'class':'jtxt'})[4]
            fsummary = summary2.findAll('span')[1].text
        except:
            fsummary = None
        dateposted1 = parent.find('div',{'class':'job_optitem ico7',
                                    'itemprop':'datePosted'})
        if dateposted1 is None:
            dateposted1 = None
        else:
            dateposted1 = dateposted1.text
        job_link = parent.find('a',{'class':'title_in'}).get('href')
        job_id = parent.get('id')
        curr_row = {'Job_title':job_title,
                    'Job_id' : job_id,
                    'Company':company,
                    'Keyskills':keyskills,
                    'Location':location,
                    'Summary':fsummary,
                    'Posted_date':dateposted1,
                    'Job_link':job_link}
        jobs = jobs.append(curr_row,ignore_index = True)
jobs
```

```python
for i, row in jobs.iterrows():
    url = row['Job_link'].replace('//','').replace(' ','%20')
    print(url)
    print('--------------------------------------------------------------------')
```

```python
jobs_detailed = pd.DataFrame()
for i, row in jobs.head(10).iterrows():
    url = row['Job_link'].replace(' ','%20')
    page = requests.get('https:'+url).text
    bs_obj = bs4.BeautifulSoup(page, 'html.parser')
    role_sec = bs_obj.find('div',{'class':'jd_rol_section'})
    span_tag = role_sec.findAll('span')
    
    industry = span_tag[0].text
    function = span_tag[1].text
    roles = '--'.join([a.text for a in span_tag[2].findAll('a')])
    posted_on = span_tag[3].text  
    curr_row = {'Job_id' : row['Job_id'],
                'Industry': industry,
                'Function': function,
                'Roles': roles,
                'Posted On':posted_on}
    jobs_detailed = jobs_detailed.append(curr_row,ignore_index = True)
jobs_detailed
```
